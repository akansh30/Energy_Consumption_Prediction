 **Household Power Consumption Dataset**

**Data Description**
This dataset contains measurements of household power consumption recorded over time, focusing on energy usage to predict future consumption patterns.

**Column Descriptions:**
**Date:** The date of the measurement in the format DD/MM/YYYY.  
**Time:** The time of the measurement in the format HH:MM:SS.  
**Global_active_power:** The total active power consumed by the household, measured in kilowatts (kW).  
**Global_reactive_power:** The total reactive power, which represents the power that does no useful work, measured in kilovolt-amperes reactive (kVAR).  
**Voltage:** The electrical potential measured at the household, in volts (V).  
**Global_intensity:** The overall current intensity flowing in the household, measured in amperes (A).  
**Sub_metering_1:** The energy consumption for kitchen appliances, measured in watt-hours (Wh).  
**Sub_metering_2:** The energy consumption for laundry appliances, measured in watt-hours (Wh).  
**Sub_metering_3:** The energy consumption for other appliances (like air conditioning), measured in watt-hours (Wh).

**Prepares the data by removing non-numeric columns, separating features and target, and splitting into training and testing sets**

**Trains a Linear Regression model, makes predictions, calculates RMSE, and computes accuracy on the test set**

**Trains a Lasso Regression model, makes predictions, calculates RMSE, and computes accuracy on the test set**

```python
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_squared_error
import numpy as np

# Initialize the Lasso Regression model
lasso_reg = Lasso()

# Train the model on the prepared dataset (df_model)
lasso_reg.fit(X_train, y_train)

# Make predictions for the test set
lasso_reg_predictions = lasso_reg.predict(X_test)

# Calculate RMSE for Lasso Regression
lasso_reg_rmse = np.sqrt(mean_squared_error(y_test, lasso_reg_predictions))

# Calculate accuracy (R-squared value)
lasso_reg_accuracy = lasso_reg.score(X_test, y_test) * 100

# Print results
print(f"Lasso Regression RMSE: {lasso_reg_rmse:.2f}")
print(f"Lasso Regression Accuracy (R^2): {lasso_reg_accuracy:.2f}%")
```

**Trains a Ridge Regression model, makes predictions, calculates RMSE, and computes accuracy on the test set**

```python
from sklearn.linear_model import Ridge
from sklearn.metrics import mean_squared_error
import numpy as np

# Initialize the Ridge Regression model
ridge_reg = Ridge()

# Train the model on the prepared dataset (df_model)
ridge_reg.fit(X_train, y_train)

# Make predictions for the test set
ridge_reg_predictions = ridge_reg.predict(X_test)

# Calculate RMSE for Ridge Regression
ridge_reg_rmse = np.sqrt(mean_squared_error(y_test, ridge_reg_predictions))

# Calculate accuracy (R-squared value)
ridge_reg_accuracy = ridge_reg.score(X_test, y_test) * 100

# Print results
print(f"Ridge Regression RMSE: {ridge_reg_rmse:.2f}")
print(f"Ridge Regression Accuracy (R^2): {ridge_reg_accuracy:.2f}%")
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit the ARIMA model
model = ARIMA(y_train, order=(5,1,0)) # Example order, you might need to tune this
model_fit = model.fit()

# Make predictions on the test set
predictions = model_fit.predict(start=len(y_train), end=len(y_train) + len(y_test) - 1, typ='levels')

# Calculate RMSE
rmse = np.sqrt(mean_squared_error(y_test, predictions))
```

### Parameters of ARIMA :

**Order of ARIMA (p, d, q)**

In our code, **order=(5,1,0)** specifies:

**p = 5**: The AR term includes 5 lagged values.

**d = 1**: The series is differenced once to make it stationary.

**q = 0**: There are no MA terms.

This means the model equation will include 5 autoregressive terms and 1 differencing step, but no moving average terms.

### **Prophet Model Overview**
The **Prophet model** is an **open-source forecasting tool** developed by **Facebook**, designed for **handling time series data** with seasonal patterns, trends, and missing values. 
It is particularly effective when the data has strong seasonality and outliers.

Prophet is **highly flexible**, allowing automatic detection of trends and seasonalities while enabling users to customize forecasts based on domain knowledge.

**Usage in the Project**
In the provided code, the Prophet model is being utilized to **forecast future values based on historical data**:

**1.Data Splitting:**
The dataset is divided into **80% for training (train_df)** and **20% for testing (test_df)** to evaluate the model's performance.

**2.Training the Model:**
The model is trained on the **training dataset (train_df)** using the **fit()** method. This step enables the model to learn patterns in the historical data.

**3.Forecasting Future Values:**

**-7-Day Prediction:** The **make_future_dataframe(periods=7)** function generates dates for the next 7 days. The model then predicts the values for these dates (predictions_7days).

**-30-Day Prediction:** Similarly, predictions for the next 30 days (predictions_30days) are generated for longer-term forecasting.

**Relevance to the Project**
In this project, the Prophet model helps to:

**a.Analyze Trends:** Identify long-term patterns in the time series data.

**b.Predict Weekly and Monthly Outcomes:** Generate short-term (7 days) and medium-term (30 days) forecasts, which are valuable 
for planning and decision-making.

**c.Handle Seasonality:** Automatically account for weekly or monthly seasonal variations in the data.

**Processes the original DataFrame df by combining the 'Date' and 'Time' columns into a single datetime column 'ds', renames the target column to 'y', and drops the original 'Date' and 'Time' columns.**

```python
# Create a copy of the original DataFrame
data = df.copy()

# Combine 'Date' and 'Time' columns into a single datetime column
data['ds'] = pd.to_datetime(data['Date'].astype(str) + ' ' + data['Time'].astype(str))

# Rename the target column for compatibility with modeling libraries like Prophet
data = data.rename(columns={'Global_active_power': 'y'})

# Drop unnecessary columns ('Date' and 'Time') after creating the 'ds' column
data.drop(['Date', 'Time'], axis=1, inplace=True)

# Print the first few rows of the processed data
print(data.head())
```

**The code splits data into training and testing sets, trains a Prophet model, generates future dates for 7 and 30 days, and then predicts values for those future dates using the trained model.**


```python
train_size = int(len(data) * 0.8)  # 80% of data for training
train_df = data[:train_size]
test_df = data[train_size:]

model = Prophet()
model.fit(train_df)

future_dates_7days = model.make_future_dataframe(periods=7) # weekly prediction
future_dates_30days = model.make_future_dataframe(periods=30) # monthly prediction

predictions_7days = model.predict(future_dates_7days)
predictions_30days = model.predict(future_dates_30days)
```
